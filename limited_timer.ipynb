This timer is set to run for 8 hours, so you can run alignments overnight.

Please resist changing it for longer periods of time so we can limit resource consumption for the Biology Hub.

```python
import time

def meaningless_meow():
    """
    Meow every 10 minutes.
    """
    
    print(f"This meow is meowed at: {time.ctime()}")
    return None

start_time = time.time()
run_time_limit = 8*60*60

if __name__ == "__main__":
    while (time.time() - start_time) < run_time_limit:
        meaningless_meow()
        time.sleep(10*60)
```
